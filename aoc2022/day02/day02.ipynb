```python
import pandas as pd

df = pd.read_csv('data.txt', header=None, names=['opponent', 'me'], sep=' ')
df['opponent'] = df.opponent.map({'A': 'rock', 'B': 'paper', 'C': 'scissors'})
df['me'] = df.me.map({'X': 'rock', 'Y': 'paper', 'Z': 'scissors'})
# (0 if you lost, 3 if the round was a draw, and 6 if you won).
df['outcome_score'] = 0
df.loc[df.me==df.opponent, 'outcome_score'] = 3
df.loc[((df.me == 'rock') & (df.opponent == 'scissors')) | ((df.me == 'paper') & (df.opponent == 'rock')) | ((df.me == 'scissors') & (df.opponent == 'paper')), 'outcome_score'] = 6
df['shape_score'] = df.me.map({'rock': 1, 'paper': 2, 'scissors': 3})
df['score'] = df.outcome_score + df.shape_score

df.head(20)
```

```python
# part one
df.score.sum()
```

```python
# part two
df = pd.read_csv('data.txt', header=None, names=['opponent', 'outcome_score'], sep=' ')
df['opponent'] = df.opponent.map({'A': 'rock', 'B': 'paper', 'C': 'scissors'})
# X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win. Good luck!"
df['outcome_score'] = df.outcome_score.map({'X': 0, 'Y': 3, 'Z': 6})
df.loc[df.outcome_score==0, 'me'] = df.opponent[df.outcome_score==0].map({'rock': 'scissors', 'paper': 'rock', 'scissors': 'paper'})
df.loc[df.outcome_score==3, 'me'] = df.opponent[df.outcome_score==3]
df.loc[df.outcome_score==6, 'me'] = df.opponent[df.outcome_score==6].map({'rock': 'paper', 'paper': 'scissors', 'scissors': 'rock'})
df['shape_score'] = df.me.map({'rock': 1, 'paper': 2, 'scissors': 3})
df['score'] = df.outcome_score + df.shape_score

df.score.sum()
```
